```python
#import libraries

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjuts the configuration pf the plots
```

```python
# Read the data from csv file

df = pd.read_csv('movies.csv')
df.head()
```

```python
# check for missing data

for col in df.columns:
    perct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, perct_missing))
```

```python
#Data types of the columns
df.dtypes
```

```python
#Changing the datatypes of columns to get rid of unnecessary decimal values

df['budget'] = df['budget'].astype('int64', errors='ignore')
df['gross'] = df['gross'].astype('int64', errors='ignore')
df['votes'] = df['votes'].astype('int64', errors='ignore')
df['runtime'] = df['runtime'].astype('int64', errors='ignore')
```

```python
df.head()
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates().head(10)
```

```python
df.head()
```

```python
# guessing : budget high correlation 
# guessing : company high correlation
```

```python
# Scatter plot with budget vs gross



plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x = 'budget' ,
            y = 'gross', 
            data = df, 
            scatter_kws={'color':'red'},
            line_kws={'color':'blue'})

```

```python
# Start looking at correlation

df.corr(method = 'pearson') #pearson , kendall, spearman
```

```python
#high correlation between budget and gross
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
